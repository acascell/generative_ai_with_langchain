```python
from langchain_community.chat_models import ChatOllama
from langchain_core.prompts import ChatPromptTemplate

# Prompt template
template = ChatPromptTemplate.from_messages([
    ("system", "You are a problem solver assistant"),
    ("user", "{problem}")
])

# Local open-source model via Ollama
chat = ChatOllama(
    model="mistral",   # or "mistral", "phi3", etc.
    temperature=0
)

# Chain
chain = template | chat

# Invoke
response = chain.invoke({
    "problem": "Explain these parameters in simpler words: “Top-k   Limits token selection to k most probable tokens   1-100   Lower values (1-10): More focused outputs Higher values: More diverse completions "
})

print(response.content)
```
